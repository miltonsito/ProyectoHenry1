```python
Netflix['platform'] = 'Netflix'
Hulu['platform'] = 'Hulu'
Disney['platform'] = 'Disney'
Amazon['platform'] = 'Amazon'
```

```python
nueva_tabla1 = pd.concat([Netflix, Hulu], axis=0)
```

```python
nueva_tabla2 = pd.concat([Amazon, Disney], axis=0)
```

```python
nueva_tabla2.head(2)
```

```python
nueva_tabla2.tail(2)
```

```python
datos_movies = pd.concat([nueva_tabla2, nueva_tabla1], axis=0)
```

```python
datos_movies.tail()
```

```python
datos_movies.isnull().sum()
```

```python
datos_movies["rating"].fillna("G", inplace=True)
```

```python
datos_movies.isnull().sum()
```

```python
datos_movies['date_added'] = pd.to_datetime(datos_movies['date_added'])
```

```python
datos_movies.head()
```

```python
datos_movies["type"] = datos_movies["type"].str.lower()
datos_movies["title"] = datos_movies["title"].str.lower()
datos_movies["director"] = datos_movies["director"].str.lower()
datos_movies["cast"] = datos_movies["cast"].str.lower()
datos_movies["country"] = datos_movies["country"].str.lower()
datos_movies["rating"] = datos_movies["rating"].str.lower()
datos_movies["duration"] = datos_movies["duration"].str.lower()
datos_movies["listed_in"] = datos_movies["listed_in"].str.lower()
datos_movies["description"] = datos_movies["description"].str.lower()
```

```python
datos_movies.head()
```

```python
datos_auxiliar = datos_movies.duration.str.split(pat=' ',expand=True)
```

```python
datos_movies["duration_int"] = datos_auxiliar[0]
```

```python
datos_movies["duration_type"] = datos_auxiliar[1]
```

```python
datos_movies["duration_type"] = datos_movies["duration_type"].str.replace("seasons", "season")
```

```python
datos_movies.tail()
```

```python
datos_movies = datos_movies.drop(['duration'], axis=1)
```

```python
datos_movies["duration_int"] = datos_movies["duration_int"].astype("Int64")
```

```python
datos_movies["duration_int"].dtype
```

```python
datos_movies = datos_movies.rename(columns={"release_year":"year"})
```

```python
datos_movies.head(1)
```

```python
datos_movies.to_csv("movies_datos2.csv", index=False)
```
